# 02 — No-Arbitrage Bounds and Put–Call Parity

Before using a pricing model, I wanted model-independent checks that any result must satisfy. This notebook works through price bounds, discounting, and put–call parity.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))
```

```python
from options_lab.arbitrage import european_option_bounds, parity_implied_put_price, put_call_parity_gap

spot, strike, rate, maturity, call = 100., 100., 0.05, 1., 10.
call_bounds = european_option_bounds(spot, strike, rate, maturity, "call")
put_bounds = european_option_bounds(spot, strike, rate, maturity, "put")
implied_put = parity_implied_put_price(call, spot, strike, rate, maturity)
pd.Series({"call_lower":call_bounds.lower,"call_upper":call_bounds.upper,
           "put_lower":put_bounds.lower,"put_upper":put_bounds.upper,
           "parity_implied_put":implied_put,
           "parity_gap":put_call_parity_gap(call, implied_put, spot, strike, rate, maturity)})
```

```python
rates = np.array([0., .02, .05, .10])
rows=[]
for r in rates:
    cb=european_option_bounds(100.,100.,r,1.,"call")
    pb=european_option_bounds(100.,100.,r,1.,"put")
    rows.append({"rate":r,"call_lower":cb.lower,"put_upper":pb.upper})
pd.DataFrame(rows)
```

**What I took from it.** A higher interest rate lowers the present value of the strike. That raises the call lower bound and lowers the European put upper bound.
