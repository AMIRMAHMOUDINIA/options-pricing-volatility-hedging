# 01 — Option Payoffs and Profit

I started here because payoff and profit are often mixed together. This notebook keeps the premium and position direction explicit and then checks that long and short positions are exact opposites under the same assumptions.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))
```

```python
from options_lab.payoffs import call_payoff, put_payoff, option_profit

spots = np.array([80., 100., 105., 120.])
strike = 100.
summary = pd.DataFrame({
    "S_T": spots,
    "call_payoff": call_payoff(spots, strike),
    "long_call_profit": option_profit(call_payoff(spots, strike), 6., "long"),
    "put_payoff": put_payoff(spots, strike),
    "long_put_profit": option_profit(put_payoff(spots, strike), 5., "long"),
})
summary
```

```python
grid = np.linspace(40, 160, 481)
profit = option_profit(call_payoff(grid, 100.), 8., "long")
plt.figure(figsize=(8,5))
plt.plot(grid, profit)
plt.axhline(0, linewidth=1)
plt.axvline(100, linestyle="--", linewidth=1)
plt.xlabel("Underlying price at expiration")
plt.ylabel("Profit per unit")
plt.title("Long Call Profit")
plt.show()
```

**What I took from it.** Finishing in the money is not enough to guarantee a profit. For the call with strike 100 and premium 6, the simplified break-even price is 106.
