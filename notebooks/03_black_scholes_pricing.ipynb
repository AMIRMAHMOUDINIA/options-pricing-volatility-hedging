# 03 — Black–Scholes Pricing

This notebook moves from the no-arbitrage checks to the Black–Scholes formula, then verifies parity and studies how the price bends as spot changes.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))
```

```python
from options_lab.black_scholes import black_scholes_d1_d2, black_scholes_price
from options_lab.arbitrage import put_call_parity_gap

d1,d2=black_scholes_d1_d2(100.,100.,.05,1.,.20)
call=black_scholes_price(100.,100.,.05,1.,.20,"call")
put=black_scholes_price(100.,100.,.05,1.,.20,"put")
pd.Series({"d1":d1,"d2":d2,"call":call,"put":put,
           "parity_gap":put_call_parity_gap(call,put,100.,100.,.05,1.)})
```

```python
spots=np.linspace(50,150,401)
calls=black_scholes_price(spots,100.,.05,1.,.20,"call")
plt.figure(figsize=(8,5))
plt.plot(spots,calls)
plt.xlabel("Current underlying price")
plt.ylabel("Call value")
plt.title("Black–Scholes Call Value versus Spot")
plt.show()
```

**What I took from it.** The option-value curve is nonlinear. Its local slope is delta and its curvature is gamma, which become the focus of the next notebook.
